#### Q1. Is there any relationship b/w Sepal_Length and Sepal_Width?

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
plt.scatter(df["Sepal_Length"], df["Sepal_Width"])
plt.title("Scatter plot of Sepal_Length vs. Sepal_Width")
plt.xlabel("Sepal_Length")
plt.ylabel("Sepal_Width")
plt.show()
```

* As we can observe in the scatter plot, the points are not scattered in any linear direction so there is no significant relationship between Sepal_Length and Sepal_Width.   

#### Q2. Plot the histogram of Petal_Length and Sepal_Width column and compare their distribution

```python
plt.hist(df["Petal_Length"])
plt.title("Histogram of Petal_Length")
plt.show()
```

```python
plt.hist(df["Sepal_Width"])
plt.title("Histogram of Sepal_Width")
plt.show()
```

#### Q3. Are there any outliers in Petal_Length column for Setosa species?

```python
plt.figure(figsize=(12,8))
plt.boxplot(df[df["Species"]=="Setosa"]["Petal_Length"])
plt.title("Boxplot of Petal_Length for Setosa species")
plt.show()
```

#### Q4. Plot the mean of Petal_Length column across all the 3 species. 

```python
df_grouped = df.groupby("Species").mean()["Petal_Length"]
df_grouped
```

```python
df_grouped = df_grouped.reset_index()
df_grouped
```

```python
plt.bar(df_grouped["Species"], df_grouped["Petal_Length"])
plt.title("Petal_Length mean across the species")
plt.xlabel("Species")
plt.ylabel("Petal_Length")
plt.show()
```
